```python
import os 
import pickle
def load_result(path):
    with open(path,'rb') as f:
        return pickle.load(f)
```

```python
# citeseer SGC 
alpha_list = [0.05,0.1,0.15,0.2,0.25]
# dataset_model = './citeseer-SGC'
# # RWalkRestart Result
# path = "".join([dataset_model, "-RWalkRestart",".pkl"])
# with open(path,'rb') as f:
#     citeseer_result_RW = pickle.load(f)
dataset_model = './citeseer-SGC'
path = "".join([dataset_model, "-RWalkRestart",".pkl"])
citeseer_result_RW = load_result(path)
path = "".join([dataset_model, "-AugNormAdj",".pkl"])
citeseer_result_AugNormAdj = load_result(path)
```

```python
citeseer_result_RW
```

```python
citeseer_result_AugNormAdj
```

```python
import matplotlib
import matplotlib.pyplot as plt
test_acc_list = []
for RW_result in citeseer_result_RW:
    test_acc_list.append(RW_result['test_acc'])
    
AugNormAdj_test_acc_list = []
for i in range(len(test_acc_list)):
    AugNormAdj_test_acc_list.append(citeseer_result_AugNormAdj['test_acc'])
plt.plot(alpha_list, test_acc_list,label='Restart')
plt.plot(alpha_list, AugNormAdj_test_acc_list,label='AugNormAdj')
plt.xlabel('alpha')
plt.ylabel('test_acc')
plt.grid()
plt.legend()

# fig.savefig("test.png")
plt.show()
```
